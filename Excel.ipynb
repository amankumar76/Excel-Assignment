Question 1: What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly explain with examples.

Answer:

'Paste' transfers all contents from the copied cell including values, formatting, formulas.
Example: If A1 contains =B1+B2 with a yellow background, using Paste will copy both the formula and formatting.

'Paste Special' allows selective pasting (values only, formulas only, formatting only).
Example: Paste Special → Values will paste only the result, not the formula.

Question 2: Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes' in Excel.

Answer:

Freeze Panes:
- Keeps specific rows/columns visible while scrolling.
- Useful for keeping headers in view.
Example: Freeze Top Row keeps column titles fixed.

Split Panes:
- Divides the worksheet into separate scrollable sections.
- Useful for comparing distant data within the same sheet.

Question 3: Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?

Answer:

Inserting a Row:
- Adds a horizontal row above the selected row.

Inserting a Column:
- Adds a vertical column to the left of the selected column.

Yes, multiple rows/columns can be inserted by selecting multiple rows/columns before clicking Insert.

Question 4: What are logical functions in Excel? Provide examples of at least two logical functions and their applications.

Answer:

Logical functions evaluate conditions and return TRUE or FALSE.

Examples:
1. IF: =IF(A1>50,"Pass","Fail")  
   Application: Grade evaluation.

2. AND: =AND(A1>10, B1<50)  
   Application: Checking multiple conditions.

Question 5: Discuss the purpose of 'XLOOKUP' and how it differs from the traditional 'VLOOKUP' function.

Answer:

The XLOOKUP function searches for a value and returns a corresponding result from another range.

Differences from VLOOKUP:
- Can look left (VLOOKUP cannot)
- Doesn’t require column index numbers
- Exact match by default
- Works both horizontally and vertically
- Less error-prone than VLOOKUP

Question 7:
Demonstrate how to insert and delete multiple rows and columns in Excel.
------------------------------------------

Answer:
To Insert Multiple Rows:
1. Select multiple rows (example: select Row 5 to Row 7).
2. Right-click → Insert.
Excel inserts the same number of new rows above the selected rows.

To Insert Multiple Columns:
1. Select multiple columns (example: select Column B and C).
2. Right-click → Insert.
New columns will be added to the left.

To Delete Multiple Rows:
1. Select the rows.
2. Right-click → Delete.

To Delete Multiple Columns:
1. Select the columns.
2. Right-click → Delete.

Use Excel's 'Find and Replace' feature to update department names in a sample table.
------------------------------------------

Answer:
Steps to use Find & Replace:
1. Press Ctrl + H.
2. In 'Find what': Type the old department name (e.g., Sales).
3. In 'Replace with': Type the new department name (e.g., Business Development).
4. Click 'Replace All'.

Excel will update all matching entries.

Question 10:
You're working with a dataset that contains missing values. As a Data Scientist, explain how
you'd detect and handle missing data using Excel.
Mention tools like:
- Go To Special
- ISBLANK
- COUNTBLANK

------------------------------------------

Answer:
Detecting Missing Data:

1. Using 'Go To Special':
   - Select the data → Press Ctrl + G → Click 'Special' → Select 'Blanks'.
   - Excel highlights all blank cells.

2. Using ISBLANK:
   Formula: =ISBLANK(A1)
   - Returns TRUE if A1 is empty, FALSE otherwise.

3. Using COUNTBLANK:
   Formula: =COUNTBLANK(A1:A50)
   - Counts total blank cells in the range.

Handling Missing Data:
- Replace blanks with 0.
- Replace blanks with average/median values.
- Replace blanks with text such as "N/A".
- Delete rows that contain too many missing values.
- Use Fill Down or Fill Series if appropriate.
